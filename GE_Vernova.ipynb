```python
# pulling data from yahoo finance
import yfinance as yf
import matplotlib.pyplot as plt
```

## Stock Performance

```python
# get the ticker symbol
ticker_symbol = 'GEV'

gev = yf.Ticker(ticker_symbol)

# get historical data from April 2, 2024 (when it was public), to April 25, 2025 
hist = gev.history(start='2024-04-02', end='2025-04-26')

hist = hist[['Open', 'High', 'Low', 'Close', 'Volume']]

print(hist.head())

# stock graph of the closing price
plt.figure(figsize=(12,6))   
plt.plot(hist.index, hist['Close'], color='green')  # x = date, y = close price
plt.title('GE Vernova (GEV) Stock Closing Price (Apr 2024 - Apr 2025)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```

```python
# get the ticker symbol and s&p 500
ticker_symbol = 'GEV'
sp500 = yf.Ticker('^GSPC')

gev = yf.Ticker(ticker_symbol)

# get historical data from April 2, 2024 (when GEV was public), to April 25, 2025) 
gev_hist = gev.history(start='2024-04-02', end='2025-04-26')
sp500_hist = sp500.history(start='2024-04-02', end='2025-04-26')

# stock graph of the Closing Price
plt.figure(figsize=(12,6))   
plt.plot(gev_hist.index, gev_hist['Close'], label='GE Vernova (GEV)', color='green')
plt.plot(sp500_hist.index, sp500_hist['Close'], label='S&P 500 Index (^GSPC)', color='blue')
plt.title('GE Vernova vs. S&P 500 (Apr 2024 - Apr 2025)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```

```python

# GEV vs. peers
# peers: ABB (ABBNY), Otis Worldwide (OTIS), CAE (CAE), Regal Rexnord (RRX), Novanta (NOVT), 
# Bloom Energy (BE), XCHG (XCH), QuantumScape (QS), Atkore (ATKR), and Energizer (ENR) 
tickers = ['GEV', 'ABBNY', 'OTIS', 'CAE', 'RRX', 'NOVT', 'BE', 'XCH', 'QS', 'ATKR', 'ENR']

data = yf.download(tickers, start='2024-04-02', end='2025-04-26')['Close']

plt.figure(figsize=(16,8))

for ticker in tickers:
    if ticker == 'GEV':
        plt.plot(data.index, data[ticker], label=ticker, color='green', linewidth=2.5)  
    else:
        plt.plot(data.index, data[ticker], label=ticker)  

plt.title('GE Vernova (GEV) vs Selected Peers (Apr 2024 - Apr 2025)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

## Top Highest Stock Prices

```python
top_10_highest_stock_prices = hist.sort_values(by='Close', ascending=False).head(10) # ascending here look at highest stock to lowest, top 10 biggest closing prices

print(top_10_highest_stock_prices[['Close']])
```

## Top Lowest Stock Prices

```python
top_10_lowest_stock_prices = hist.sort_values(by='Close', ascending=True).head(10) # ascending here look at lowest stock to highest, bottom 10 smallest closing prices

print(top_10_lowest_stock_prices[['Close']])
```

## Valuation Multuples
